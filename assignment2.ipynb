```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import silhouette_score
```

```python
X, _ = make_circles(n_samples=300, noise=0.1, factor=0.5)
```

```python
plt.scatter(X[:, 0], X[:, 1], c='b', marker='o', s=30)
plt.title("Synthetic Circular Data")
plt.show()
```

```python
neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.plot(distances)
plt.title("Elbow Method to Determine K")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans_labels = kmeans.fit_predict(X)
```

```python
gmm = GaussianMixture(n_components=2)
gmm_labels = gmm.fit_predict(X)
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
spectral = SpectralClustering(n_clusters=2, eigen_solver='arpack', affinity="nearest_neighbors")
spectral_labels = spectral.fit_predict(X)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(141)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='rainbow', marker='o', s=30)
plt.title("KMeans Clustering")

plt.subplot(142)
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='rainbow', marker='o', s=30)
plt.title("GMM Clustering")

plt.subplot(143)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='rainbow', marker='o', s=30)
plt.title("DBSCAN Clustering")

plt.subplot(144)
plt.scatter(X[:, 0], X[:, 1], c=spectral_labels, cmap='rainbow', marker='o', s=30)
plt.title("Spectral Clustering")

plt.show()
```

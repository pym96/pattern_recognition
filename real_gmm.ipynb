```python
from sklearn import datasets
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
data = iris.data
data
```

```python
# Apply GMM clustering
n_components = 3
gmm = GaussianMixture(n_components=n_components)
gmm.fit(data)
labels = gmm.predict(data)
```

```python
# Reduce dimensions for visulization
pca = PCA(n_components=2).fit_transform(data)
```

```python
# Visualize the clusters
plt.figure(figsize=(8,6))
for i in range(n_components):
    plt.scatter(pca[labels == i, 0], pca[labels == i, 1], label=f'Cluster {i + 1}', edgecolors='k')
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], s=300, c='red', marker="*",label='Centroids')
plt.legend()
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('GMM Clustering on Iris Dataset')
plt.show()
```

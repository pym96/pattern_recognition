```python
import cv2 as cv
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load image
image_path = './panda.jpg'
image = cv.imread(image_path)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # Convert to RGB for proper display using matplotlib
plt.imshow(image_rgb)
plt.title('Original Image')
plt.show()

# Reshape the image to be a list of pixels and normalize
image_pixels = image.reshape((-1, 3))
_mean = np.mean(image_pixels, axis=0, keepdims=True)
_std = np.std(image_pixels, axis=0, keepdims=True)
normalized_pixels = (image_pixels - _mean) / _std

# Number of clusters
ncomp = 4

# KMeans Initialization
kmeans = KMeans(n_clusters=ncomp, random_state=0).fit(normalized_pixels)
initial_centers = kmeans.cluster_centers_

# GMM
gmm = GaussianMixture(n_components=ncomp, init_params='kmeans', means_init=initial_centers).fit(normalized_pixels)
labels = gmm.predict(normalized_pixels)

# Visualize the segmentation
segmented_image = np.zeros_like(image)
colors = np.array([
    [255, 0, 0],
    [0, 255, 0],
    [0, 0, 255],
    [255, 255, 0]
])

# Assuming your original image shape was (height, width, channels)
height, width, _ = segmented_image.shape

# Reshape labels to be 2D
labels_2d = labels.reshape(height, width)

for i in range(ncomp):
    segmented_image[labels_2d  == i] = colors[i]

segmented_rgb = cv.cvtColor(segmented_image, cv.COLOR_BGR2RGB)
plt.imshow(segmented_rgb)
plt.title('Segmented Image')
plt.show()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X, _ = make_circles(n_samples=1000, noise=0.1, factor=0.1)
Y, _ = make_circles(n_samples=1000, noise=0.1, factor=0.5)
```

```python
plt.scatter(X[:,0], X[:, 1], c='b', marker='o',s=20)
```

```python
plt.scatter(Y[:,0], Y[:, 1], c='b', marker='o',s=20)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
distances = np.sort(distances, axis=0) # Getting distance from different clusters
distances = distances[:,1]
plt.plot(distances)
```
